```python
# The parsing of our “Untitled.xml” file starts at the root of the tree
# namely the <data> element, which contains the entire data structure.
xtree = et.parse("Untitled.xml")
xroot = xtree.getroot()
```

```python
# check the rep name for loreal or wella or unknown xml file
def get_supplier(root):
    for item in xroot.findall("./rep/name"):
        if item.text == 'Wella':
            supplier = 'Wella'
        elif item.text == 'Loreal':
            supplier = 'Loreal'
        else:
            supplier = 'Unknown'
    return supplier
supplier = get_supplier(xroot)
print(supplier)
```

```python
def get_stock_qty(root):
    stock_no = []
    qty = []
    for item in xroot.iter('stock'):
        stock_no.append(item.text)
    for item in xroot.iter('quantity'):
        qty.append(item.text)
    return stock_no, qty
        
stock_no, qty = get_stock_qty(xroot)
```

```python
# Check the size of the two lists
print(len(qty),len(stock_no))
if len(qty) != len(stock_no):
    print("Mismatch qty of items and PNs ")
```

```python
#convert lists into dictionart , step needed to create dtatframe
order_dict = dict(zip(stock_no, qty))
print(order_dict)
```
